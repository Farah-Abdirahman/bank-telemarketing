```python
import pandas #polars library
data = pandas.read_csv('https://modcom.co.ke/datasets/bank_full.csv')
data.head(5)
```

```python
# Exploratory Data Analysis (EDA) is a fundamental step in data science projects.
# It involves studying and exploring datasets to understand their key characteristics, discover patterns, identify outliers, and establish relationships between variables.
# Here are the primary goals of EDA:

# 1. Data Cleaning:

# EDA examines data for errors, missing values, and inconsistencies.
# Techniques like data imputation, handling missing data, and outlier detection are employed.

# 2. Descriptive Statistics:

# EDA uses statistical measures (e.g., mean, median, standard deviation, range) to understand variable tendencies, variability, and distributions.
```

```python

data.shape
```

```python
#ML rules..
#1. No Empties.-Does not work with empties - remove empties or fill them(Imputation process)
#2. Ml does not work with text - convert text to digits - process called (Encode data)
#Check Empties
data.isnull().sum()#Checks empties
```

```python
#Fill empties
data['job'].fillna('unknown', inplace = True)
data['marital'].fillna('unknown', inplace = True)
data['loan'].fillna('unknown', inplace = True)


#For duration is a continous variable. We find the mean/median

duration_mean = data['duration'].mean()
duration_mean
data['duration'].fillna(duration_mean, inplace=True)
data.isnull().sum()#Checks empties
# For dropping records use dropna
```

```python
# Ordinal encoding assigns integer values to each category of a categorical variable based on their order or hierarchy in the dataset.
# In ordinal encoding, the categories are replaced with integers starting from 0 or 1, based on their order.
data['loan'].replace({'no':0, 'yes':1, 'unknown':2},inplace=True)
data['contact'].replace({'telephone':0, 'cellular':1},inplace=True)
data['default'].replace({'no':0, 'yes':1, 'unknown':2},inplace=True)
data['housing'].replace({'no':0, 'yes':1, 'unknown':2},inplace=True)
data['marital'].replace({'divorced':0, 'married':1, 'single':2,'unknown':3},inplace=True)
data['poutcome'].replace({'nonexistent':0, 'failure':1, 'success':2},inplace=True)

data.tail(10)
```

```python
def slice(param):
  if '-' in param:
    answer = param.split('-')
    return answer[1]

  else:
    return param




```

```python
#Label Encoding is a technique that is used to convert categorical columns into numerical ones
# so that they can be fitted by machine learning models which only take numerical data
#Education, Month , Week, Job
# dict vectorizer -  add it to the presentation
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['job'] = encoder.fit_transform(data['job'])
data['education'] = encoder.fit_transform(data['education'])
data['month'] = encoder.fit_transform(data['month'])
data['day+AF8-of+AF8-week'] = encoder.fit_transform(data['day+AF8-of+AF8-week'])



data['cons.conf.idx'] = data['cons.conf.idx'].apply(slice)
data['emp.var.rate'] = data['emp.var.rate'].apply(slice)

data.tail()
```

```python
data.groupby('emp.var.rate').size()
```

```python
array = data.values
array
X = array[:,0:20]

Y = array[:,20]
print(X)
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y, test_size=0.3,random_state=42)
len(X_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
myoption = GradientBoostingClassifier()
myoption.fit(X_train,Y_train)
```

```python
predictions = myoption.predict(X_test)
print('Predicted Results', predictions)
print('Expected Outcome', Y_test)
```

```python
from sklearn.metrics import accuracy_score
print('The accuracy is',accuracy_score(predictions,Y_test))
```

```python
#Test Results
from sklearn.metrics import classification_report
print(classification_report(predictions,Y_test))
```

```python
#Total test data was 12357
#No were 11310 and yes were 1047
#From matrix it got 10644 right for no's and 681 wrong
#From yes it got 708 right and 324 wrong
#Model did not perform well in yes
#Possible reasons: Data Imbalance
#Another reasons: Some columns might not be necessary
from sklearn.metrics import confusion_matrix
print(confusion_matrix(predictions,Y_test))
```
